```python
from stable_baselines3 import DQN
import gymnasium as gym

env=gym.make("CartPole-v1")
model_dqn=DQN(policy='MlpPolicy'
              ,env=env,
              learning_rate=1e-3
              ,gamma=0.9
              ,buffer_size=50000  #storage of experiences
              ,batch_size=32  #takes samples as mini-batch 
              ,learning_starts=1000,)
```

```python
model_dqn.learn(total_timesteps=100_000)
```

```python
from stable_baselines3 import PPO
model_ppo=PPO(policy='MlpPolicy',learning_rate=1e-3,env=env,gamma=0.99,verbose=1)
model_ppo.learn(total_timesteps=100_000)
```

```python
def evaluate_model(model,env,episodes=10):
    rewards=[]
    for episode in range(episodes):
        done=False
        obs,_=env.reset()
        total_reward=0
        while not done:
            action,_=model.predict(obs)
            obs,reward,terminated,truncated,_=env.step(action)
            done=terminated or truncated
            total_reward+=reward
        rewards.append(total_reward)
    return sum(rewards)/len(rewards)
```

```python
print("Average reward of DQN:",evaluate_model(model_dqn,env))
print("Average reward of PPO:",evaluate_model(model_ppo,env))
```
