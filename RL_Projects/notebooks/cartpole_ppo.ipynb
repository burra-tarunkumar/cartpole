```python
from stable_baselines3 import PPO
import gymnasium as gym
train_env =gym.make('CartPole-v1')
```

```python
model_ppo=PPO(policy="MlpPolicy",
              env=train_env,
              learning_rate=1e-3,
              gamma=0.99,
              verbose=1)
model_ppo.learn(total_timesteps=100000)
```

```python
def evaluate_model(model,env,episodes=10):
    
    rewards=[]
    for _ in range(episodes):
        obs,_=train_env.reset()
        done=False
        total_rewards=0
        while not done:
            action,_=model_ppo.predict(obs,deterministic=True)
            obs,reward,terminated,truncated,_=train_env.step(action)
            done=terminated or truncated
            total_rewards+=reward
        rewards.append(total_rewards)
    return sum(rewards)/len(rewards)
```

```python
print(f"avg reward:{evaluate_model(model_ppo,train_env)}")
```

###Saave the model only once that too after the reward is considerable

```python
""" Since the average reward is greater than 200 , we can save the model """
model_ppo.save('ppo_cartpole')
```
